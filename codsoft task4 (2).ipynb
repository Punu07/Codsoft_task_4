## Spam sms detection

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\punam\Downloads\archive (8)\spam.csv",encoding ='latin1')
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
l_e = LabelEncoder()
```

```python
l_e
```

```python
df['label'] = l_e.fit_transform(df['v1'])
```

```python
x = df['v2']
x
```

```python
y= df['label']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfid = TfidfVectorizer()
```

```python
tfid
```

```python
x_tr_tf = tfid.fit_transform(x_train)
```

```python
x_te_tf = tfid.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb = MultinomialNB()
```

```python
nb
```

```python
nb.fit(x_tr_tf,y_train)
```

```python
y_pred_nb = nb.predict(x_te_tf)
```

```python
from sklearn.metrics import classification_report,roc_auc_score
```

```python
classification_report(y_test,y_pred_nb)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr
```

```python
lr.fit(x_tr_tf,y_train)
```

```python
y_Pred_lr = lr.predict(x_te_tf)
```

```python
classification_report(y_test,y_Pred_lr)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(probability = True,random_state = 42)
```

```python
svm.fit(x_tr_tf,y_train)
```

```python
y_pred_svm = svm.predict(x_te_tf)
```

```python
classification_report(y_test,y_pred_svm)
```

```python
roc_auc_score(y_test,y_pred_nb)
```

```python
roc_auc_score(y_test,y_Pred_lr)
```

```python
roc_auc_score(y_test,y_pred_svm)
```
